# Тема “Вычисления с помощью Numpy”

### Exercise 1

```python
import numpy as np
a = np.array([[1, 2, 3, 3, 1],[6, 8, 11, 10, 7]]).T
print("a.shape: {}".format(a.shape))
print("a: {}".format(a))

mean_a = a.mean(axis=0)
print("mean_a: {}".format(mean_a))
```

### Exercise 2

```python
a_centered = a - mean_a
print("a_centered: {}".format(a_centered))
print("a_centered.shape: {}".format(a_centered.shape))
```

### Exercise 3

```python
a_centered_sp = a_centered[:,0].dot(a_centered[:,1])
print("a_centered_sp: {}".format(a_centered_sp))

a_centered_sp /= a_centered.shape[0] - 1
print("a_centered_sp: {}".format(a_centered_sp))
```

# Тема “Работа с данными в Pandas”

### Exercise 1

```python
import pandas as pd

authors = pd.DataFrame({
    "author_id": [1, 2, 3],
    "author_name":  ['Тургенев', 'Чехов', 'Островский']
})
print(authors)

book = pd.DataFrame({
    "author_id": [1, 1, 1, 2, 2, 3, 3],
    "book_title": ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
    "price":  [450, 300, 350, 500, 450, 370, 290]
})
print(book)
```

### Exercise 2

```python
authors_price = pd.merge(authors, book, on='author_id', how='inner')
authors_price
```

### Exercise 3

```python
top5 = authors_price.sort_values(by='price', ascending=False).head()
top5
```

### Exercise 4

```python
groupby = authors_price.groupby("author_name")
authors_stat = groupby.agg({"price": ["min", "max", "mean"]})
authors_stat.columns = ["_".join(x[::-1]) for x in authors_stat.columns.ravel()]
authors_stat
```

### Exercise 5

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']

book_info = pd.pivot_table(authors_price, 
                   values='price', index=['author_name'],
                   columns=['cover'], aggfunc=np.sum)
book_info = book_info.fillna(0)
print(book_info)

book_info.to_pickle("book_info.pkl")
book_info2 = pd.read_pickle("book_info.pkl")
print(book_info2)

if book_info.equals(book_info2):
    print('dataframes equals')
```
