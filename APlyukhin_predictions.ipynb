```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Получение тренировочных данных

```python
df_train = pd.read_csv('train.csv')
```

### Подготовка тренировочных данных

```python
df_train_dummy = df_train.drop(columns=['Healthcare_1','LifeSquare','Id'])
df_train_dummy = pd.get_dummies(df_train_dummy)

X = df_train_dummy.drop(columns=['Price'])
y = df_train_dummy['Price']

from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.20, random_state=42)
```

### Обучение модели

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=10, random_state=42)

model.fit(X_train, y_train)
y_pred = model.predict(X_valid)

from sklearn.metrics import r2_score
r2 = r2_score(y_valid, y_pred)
r2
```

### Получение тестовых данных

```python
df_test = pd.read_csv('test.csv').set_index('Id')
```

### Получение результирующего столбца

```python
df_test_dummy = df_test.drop(columns=['Healthcare_1','LifeSquare'])
df_test_dummy = pd.get_dummies(df_test_dummy)

X = df_test_dummy
y_pred = model.predict(X)
y_pred
```

### Сохранение результатов

```python
s_pred = pd.Series(y_pred, index=X.index, name='Price')
s_pred.to_csv('s_pred.csv')
```
